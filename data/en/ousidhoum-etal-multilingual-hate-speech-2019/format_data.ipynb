```python
import pandas as pd
df = pd.read_csv('en_dataset.csv')
print(df.groupby(['sentiment'])['tweet'].count())
def rearange_class(x):
    if 'hateful' in x:
        return 1
    elif x == 'normal':
        return 0
    else:
        return 2
df.rename(columns={'sentiment':'hate','HITId':'text_id','tweet':'text'},inplace=True)
df.drop(['directness','annotator_sentiment','target','group'],axis=1,inplace=True)
df['hate'] = df.apply(lambda x: rearange_class(x.hate),axis=1)
df['text_id'] = df.apply(lambda x: 'ousidhoum_'+str(x.text_id),axis=1)
# 0 - neither
# 1 - hate speech
# 2 - offensive
df['hate'] = df['hate'].astype(int)
total = df.shape[0]
hate = df.groupby(['hate'])['text_id'].count()[1]
offensive = df.groupby(['hate'])['text_id'].count()[2]
print("Size of dataset:", total)
print("Percentage abusive:", hate/total)
print("Percentage offensive:",offensive/total)
print('Writing to file, ousidhoum.csv')
df.to_csv('ousidhoum.csv',index=False)
```
